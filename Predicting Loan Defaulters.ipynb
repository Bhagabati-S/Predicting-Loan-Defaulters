```python
from scipy.stats import ttest_ind

# perform t-test between sanctioned and disbursed amount of secondary loans
secondary_loans = data[data['PRI.NO.OF.ACCTS'] > 0]
sanctioned_secondary = secondary_loans['PRI.SANCTIONED.AMOUNT']
disbursed_secondary = secondary_loans['PRI.DISBURSED.AMOUNT']
t, p = ttest_ind(sanctioned_secondary, disbursed_secondary)

# print results
print('T-Statistic: ', t)
print('P-value: ', p)
```

```python
from scipy.stats import f_oneway

# perform ANOVA between sanctioned and disbursed amount of loans across different employment types
employed = data[data['Employment.Type'] == 'Salaried']
self_employed = data[data['Employment.Type'] == 'Self employed']
t, p = f_oneway(employed['PRI.SANCTIONED.AMOUNT'], employed['PRI.DISBURSED.AMOUNT'], 
                self_employed['PRI.SANCTIONED.AMOUNT'], self_employed['PRI.DISBURSED.AMOUNT'])

# print results
print('F-Statistic: ', t)
print('P-value: ', p)
```
